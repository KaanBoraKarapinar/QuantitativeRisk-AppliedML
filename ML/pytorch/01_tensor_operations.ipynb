## Tensor shape and dim

```python
t = torch.zeros(6)
print(t)
print(t.dim()) # How many dimensions does my tensor have? / rank of my tensor
```

```python
t = torch.zeros(2,3) #standart vector notation, [y,x]
print(t)
print(t.dim())
print(t.shape)
```

```python
t = torch.zeros(2,5,6)
print(t)
print(t.dim())
```
