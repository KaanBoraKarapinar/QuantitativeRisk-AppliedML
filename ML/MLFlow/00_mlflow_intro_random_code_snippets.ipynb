### Components of MlFlows

1. **MLflow Tracking**
   Records metrics and parameters from training runs, lets you query experiment data, and stores models, artifacts, and code. This is the part most relevant to a hyperparameter loop like the KNN example.
2. **MLflow Models**
   Standardizes trained models into a common format for deployment. Lets you load models from different frameworks (sklearn, xgboost, pytorch, etc.) through a consistent interface, and supports building customized model flavors.
3. **Model Registry**
   Stores and versions ML models. Manages a model's lifecycle (staging, production, archived) and lets you load or deploy a specific version.
4. **MLflow Projects**
   Packages ML code along with its environment (conda/pip dependencies) so the project is reproducible and repeatable by anyone, including future you.


### Experiments and Runs

An **experiment** is the top-level container in Tracking. Think of it as a folder that groups related runs together (e.g. "KNN Hyperparameter Tuning").

A **run** is a single execution within that experiment, for example one specific `n_neighbors` value with its logged parameters and metrics.

#### Mlflow Clients vs module

- **`mlflow.*` (module-level API)**: Simple, linear workflows. You typically set an experiment once at the top of the script and continue from there.
- **`MlflowClient`**: A more programmatic, lower-level interface, useful when managing multiple experiments or needing finer control.
Both accomplish the same things, just with different styles. For a straightforward loop, `mlflow.*` is the simpler starting point.

#### Key Operations

| Action | MLflow Client | MLflow Module |
|---|---|---|
| Create experiment | `client.create_experiment("Name")` | `mlflow.create_experiment("Name")` |
| Tag experiment | `client.set_experiment_tag("Name", k, v)` | `mlflow.set_experiment_tag(k, v)` |
| Delete experiment | `client.delete_experiment("Name")` | `mlflow.delete_experiment("Name")` |
| Set active experiment | (n/a) | `mlflow.set_experiment("Name")` |

#### Starting a New Experiment

```python
import mlflow

# Create new experiment
mlflow.create_experiment("My Experiment")

# Tag new experiment
mlflow.set_experiment_tag("scikit-learn", "lr")

# Set the experiment
mlflow.set_experiment("My Experiment")
```

# LOG PARAM VS LOG METRICS

* log_param vs log_metric: hyperparameters (lr, optimizer type, architecture choices) are log_param — fixed for the run. Things that change over training (loss per epoch, val RMSE) are log_metric, usually called in your training loop with a step argument.
