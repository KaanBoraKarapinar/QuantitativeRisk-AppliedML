```python
import plotly.graph_objects as go
```

Plotly has

-> figure object (go.Figure)

-> multiple traces: data object
    fig.add_trace
        and as a trace type go.Scatter

-> layout: customization of view


```python
# import plotly.graph_objects as go


x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 17, 20]

fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x=x,
        y=y,
        mode='lines',
        name='Example'
    )

)

#LAYOUT PART:
fig.update_layout(
    title='Example Graph',
    xaxis_title='X Axis',
    yaxis_title='Y Axis'
)

fig.show()
```

## Multiple Traces in Same Graph

```python
#import plotly.graph_objects as go


x = [1, 2, 3, 4, 5]
y1 = [10, 15, 13, 17, 20]
y2 = [8, 12, 15, 14, 18]

fig = go.Figure()

fig.add_trace(go.Scatter(x=x, y=y1, mode='lines', name='Seri A'))
fig.add_trace(go.Scatter(x=x, y=y2, mode='lines', name='Seri B'))

fig.show()
```

# Make Subplots

```python
from plotly.subplots import make_subplots

x = [1, 2, 3, 4, 5]
y1 = [10, 15, 13, 17, 20]
y2 = [5, 8, 6, 9, 7]


fig = make_subplots(
    rows=1, cols=2,
    subplot_titles=('Left Panel', 'Right Panel')
)

## TRACE PART

fig.add_trace(
    go.Scatter(x=x, y=y1, mode='lines', name='Seri A'),
    row=1, col=1 # LOCATION IS DEFINED HERE,
)

fig.add_trace(
    go.Scatter(x=x, y=y2, mode='lines', name='Seri B'),
    row=1, col=2
)

fig.update_layout(
    title='Two Panel Example',
    width=1000,
    height=400
)
```

# Using Loops for Graph Generation

```python
import numpy as np
import plotly.graph_objects as go

np.random.seed(1)
x = list(range(1, 21))  # 20 nokta

x = [1,2,3,4,5]

series_data = {
    'stock_nr1': [8, 12, 15, 14, 18],
    'stock_nr2': [10, 15, 13, 17, 20],
    'stock_nr3': [5, 8, 6, 9, 7],
    'stock_nr4': [3, 4 , 7, 9, 11],
}

colors = ['#FF3131', '#00FFFF', '#39FF14', '#FF00FF', '#FFD700']

fig = go.Figure()

for (label, y), color in zip(series_data.items(), colors):
    fig.add_trace(
        go.Scatter(x=x, y=y, mode='lines', name=label, line=dict(color=color))
    )

fig.show()
```

# Layout Elements

### Dropdown Menu with updatemenus

```python
import plotly.graph_objects as go

x = [1, 2, 3, 4, 5]
fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=[10,15,13,17,20], name='Seri A'))
fig.add_trace(go.Scatter(x=x, y=[5,8,6,9,7], name='Seri B'))

fig.update_layout(
    updatemenus=[
        dict(
            type="buttons",
            direction="right",
            buttons=[
                dict(label="Tümü",
                     method="update",
                     args=[{"visible": [True, True]}]),
                dict(label="Sadece A",
                     method="update",
                     args=[{"visible": [True, False]}]),
                dict(label="Sadece B",
                     method="update",
                     args=[{"visible": [False, True]}]),
            ],
            x=0.1, y=1.15
        )
    ]
)
fig.show()
```
