## apply vs map
With map(), the function receives one cell at a time — it has no awareness of the row or column it came from:

```python
basic_df = pd.DataFrame({
    "A": [1, 2, 3],
    "B": [4, 5, 6]
})

result = basic_df.map(lambda x: x * 2)
# x = 1, then 4, then 2, then 5, then 3, then 6
# completely isolated scalar each time
result
```

With apply(axis=1), the function receives an entire row as a Series:

```python
result = basic_df.apply(lambda x: x["A"] + x["B"], axis=1)
# x = [1, 4], then [2, 5], then [3, 6]
# x is a full row, you index into it by column name
result
```
